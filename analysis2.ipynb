```python
import pandas as pd # for data analysis
import numpy as np # for numerical functions
import plotly.express as px #for interactive plots
```

```python
df = pd.read_json('flipkart_fashion_products_dataset.json')
df.head()
```

```python
df.info()
```

```python
df['actual_price'].str.replace(',','')
```

```python
def clean_price(value):
    if not value:
        return np.nan
    elif isinstance(value,str):
     value = value.replace(',','')
    if value.isnumeric():
        return float(value)
    else:
     return value
    df['actual_price'].apply(clean_price)
```

```python
df['actual_price'] = df['actual_price'].apply(clean_price)
df['selling_price'] = df['selling_price'].apply(clean_price)
df.info()
```

```python
df.info()
```

```python
def clean_discount(value):
    if not value:
        return np.nan
    elif isinstance(value,str):
        value =value.replace('%.off','')
        try:
            value = float(value)
            return value
        except:
            print(value)
            return np.nan
        else:
            return value
        
```

```python
df['discount'].apply(clean_discount)
```

```python
def clean_rating(value):
    if value:
        return float(value)
    return np.nan
```

```python
df['average_rating'].apply(clean_rating)
```

```python
df['average_rating'] = df['average_rating'].apply(clean_rating)
```
